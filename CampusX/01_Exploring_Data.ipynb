# Bivariate Analysis

```python
tips = pd.read_csv("tips.csv")
print(tips.shape)
tips.sample(2)
```

```python
flights = pd.read_csv("flights.csv")
print(flights.shape)
flights.sample(2)
```

```python
iris=pd.read_csv("Iris_data.csv")
print(iris.shape)
iris.sample(2)
```

```python
titanic = pd.read_csv("train.csv")
print(titanic.shape)
titanic.sample(2)
```

### 1. Sactterplot (Numerical-Numerical)

```python
sns.scatterplot(x=tips['total_bill'],y=tips['tip'],hue=tips['sex'])
```

### 2. Barplot (Numerical-Categorical)

```python
sns.barplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

```python
sns.boxplot(x=titanic['Sex'],y=titanic['Age'],hue=titanic['Pclass'])
```

```python
sns.boxplot(x=titanic['Pclass'],y=titanic['Age'],hue=titanic['Sex'])
```

### 4. Distplot (Numerical-Categorical)

```python
sns.distplot(titanic[titanic['Survived']==1]['Age'],hist=False)
sns.distplot(titanic[titanic['Survived']==0]['Age'],hist=False)
```

### 5. Heatmap (Categorical-Categorical)

```python
sns.heatmap(pd.crosstab(titanic['Survived'],titanic['Pclass']))
```

### 6. Pairplot

```python
titanic.head(1)
```

```python
sns.pairplot(titanic[['Pclass','Fare']])
```

```python
sns.pairplot(titanic[['Age', 'Fare', 'Survived']], hue='Survived')
```

```python
sns.pairplot(iris.iloc[:,1:],hue='Species')
```

### 7. Lineplot (Numerical-Numerical)

```python
flights.sample(2)
```

```python
new= flights.groupby('year').sum().reset_index()
new.head()
```

```python
sns.lineplot(x=new['year'],y=new['passengers'])
```

```python
flights['month'].unique()
```

```python
new_v1=flights.pivot_table(values='passengers',index='month',columns='year')

month_order = [
    'January', 'February', 'March', 'April', 'May', 'June', 'July',
       'August', 'September', 'October', 'November', 'December'
]

new_v1 = new_v1.reindex(month_order)
new_v1.head()
```

```python
sns.heatmap(new_v1)
```
